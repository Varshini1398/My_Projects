```python
# to find elastic or not 
def ped (p1,p2,q1,q2):
    Q=((q2-q1)/(q2+q1)/2)
    P=((p2-p1)/(p2+p1)/2)
    PED=round(Q/P,2)
    print(f'PED: {PED}')
    if abs(PED)>1:   # abs is absolute
      print("elastic demand")
    else:
      print("Inelastic demand")
```

```python
ped(40,30,50,65)
```

```python
def ped (p1,p2,q1,q2):
    Q=((q2-q1)/(q2+q1)/2)
    P=((p2-p1)/(p2+p1)/2)
    PED=round(abs(Q/P,2))
    print(f'PED: {PED}')
    if (PED)>1:   # how to not get negative answers check after class
      print("elastic demand")
    else:
      print("Inelastic demand")
```

```python
ped(40,30,50,65)
```
